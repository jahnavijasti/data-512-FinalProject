# World Unemployment Rate

## Abstract


People, educated and uneducated, who do not work and cannot earn enough for their own living, are called unemployed. The unemployment ratio of a country is calculated by dividing the number of unemployed people by the total number of employed people in the country.

However, unemployment rate is calculated in different ways in different countries. Some of the ways in which the rate is calculated:

-- count insured unemployed only

-- count people that are receiving lfare benefit

-- disabled/ permanently unemployable people only

-- count people who choose not to work only, etc.

Understanding the meaning of unemployment is a big task. People such as retirees, students cannot be group into unemployment. A person who is eligible to work, available for work and is actively seeking for a job from four weeks is considered as unemployed. This definition is only for full-time workers. Part time workers who are seeking for full time are still not grouped into unemployed. 

Unemployment is not just one indicator on how the economy is doing. In the country like US, there are five other monthly measures of labor underutilization: labor force participation rates, employment-population ratios, average weekly wages, average hours worked. A team called Current Population Survey (CPS) asks people questions like whether they are working or searching for work, how long have then been without a job, why have they not been looking for job. All the answers from these questions are used to calculate other measures of underutilization. 

There is also a factor called Youth unemployment. Like general unemployment, youth unemployment is also usually recorded to study the job market in a country level and also worldwide. Individuals who are between 15 and 24 years of age, and who are eligible to work,  come under youth unemployees. Usually, youth unemployment is more than general unemployment because it is difficult for a fresh graduate to find a job immediately, when compared to an experienced person. 

For the purpose of this project, values published by the World Bank Data, collected from the International Labor Organization database, is used. The ILO calculates the unemployment rate by counting only the people who are currently not working but are willing to and are able to work for pay.

## Findings

_Research Question1_ and _Question 2_ is to find the countries that have the highest and lowest mean unemployment rate from 2010 to 2017. So, first let's make a new dataframe _data2010_ that has all the country's data from 2010 to 2017. Then the mean for each of the country is calculated and added to the dataframe as a new column.

Then the top and bottom 15 countries are extracted into a new dataframe. 

### Mean from 2010-2017

```python
data2010 = data.drop(data.columns[[range(1,20)]], axis=1) 

data2010["mean"] = data2010.mean(axis=1)

l15 = data2010.sort_values("mean").head(15)

h15 = data2010.sort_values("mean").tail(15)

df15 = pd.concat([l15,h15])
```

Once we have the dataframe with 30 countries, that has 15 highest and 15 lowest mean unemployment rates, we plot them.

```python
x = np.arange(30)
y = df15['mean']
my_xticks = df15['Country Name']

fig = plt.figure(figsize=(10,10))
plt.plot(x,y)
plt.xticks(x, my_xticks, rotation=90)
plt.yticks(np.arange(y.min(), y.max(), 1))
plt.grid(axis='y', linestyle='-')
plt.title('Mean')
plt.show()
##fig.savefig('/Users/tondapu/Desktop/HCDE Final Project/change.png')
```

From the above plot, we can see that countries like Cambodia, Qatar, Belarus, Thailand has mean unemployment rate less than 1%, while countries like Spain, Greece and South Africa have more than 20%. 

### Change in rate from 2010 to 2017

_Research Questions 3_ and _4_ is to find the countries which have an increase/decrease in the  rate of change in the unemployment rate. 

So, we first create a dataframe that data from 2010 to 2017. Then find the percent change in the rate between these years, and add it to the dataframe as a new column and then plot it.

```python
data2010 = data.drop(data.columns[[range(1,20)]], axis=1) 

data2010["change"] = ((data2010['2017'].values - data2010['2010'].values)/data2010['2010'].values)*100

l15 = data2010.sort_values("change").head(15)

h15 = data2010.sort_values("change").tail(15)

df15 = pd.concat([l15,h15])
```

```python
x = np.arange(30)
y = df15['change']
my_xticks = df15['Country Name']

fig = plt.figure(figsize=(10,10))
plt.plot(x,y)
plt.xticks(x, my_xticks, rotation=90)
plt.yticks(np.arange(y.min(), y.max(), 10))
plt.grid(axis='y', linestyle='-')
plt.title('Mean')
plt.show()
fig.savefig('/Users/tondapu/Desktop/HCDE Final Project/%change.png')
```

This plot only shows 15 countries that have high increase in the rate and 15 countries that have high decrease in the rate. 

Maldives, Hungary have reduced their unemployment rate by 75 and 55% respectively, in the last 7 years. Seeing this percent decrease in the unemployment, it is very encouraging.

However, countries like Greece and Cyprus has an increase in the rate over 65%. These countries really need to look into the issues that are causing this hike.
